# Линейная регрессия: аналитический и численный подходы

Сгенерируем данные для задачи регрессии

```python
import numpy as np
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt

rng = np.random.RandomState(42)

X, y = make_regression(n_samples=100, n_features=1, random_state=0, noise=10.0, bias=100.0)
```

```python
plt.scatter(X, y);
```

## Аналитический подход

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
pred_lr = lr.predict(X)
```

Подсчитаем качество предсказания (будем использовать метрику $R^2$) и визуализируем результат.

```python
from sklearn.metrics import r2_score

plt.scatter(X, y, c='blue')
plt.plot(X, pred_lr, '-', c='green')

print('r2:', r2_score(y, pred_lr))
```

## Численный подход
Теперь обучим SGDRegressor и сделаем предсказание

```python
from sklearn.linear_model import SGDRegressor

sgd = SGDRegressor()
sgd.fit(X, y)

pred_sgd = sgd.predict(X)
```

Добавим предсказание на график с исходными объектами и выxbckbv $R^2$.

```python
plt.scatter(X, y, c='blue')
plt.plot(X, pred_sgd, '-', c='red')

print('r2:', r2_score(y, pred_sgd))
```

Сравним качество предсказаний точным и численным методами

```python
print('r2 exact  :', r2_score(y, pred_lr))
print('r2 numeric:', r2_score(y, pred_sgd))
```

Численный подход дает чуть меньшее значение метрики качества, но разница в значениях  R2 
начинается в шестом знаке после запятой, то есть можно смело считать, что разницы в качестве нет.
