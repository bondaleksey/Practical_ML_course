Найдем локальный минимум функции $f(x) = x^4 - 10x + 3$

Будем искать минимум с гиперпараметрами:

$x_0 = 5$, $\eta = 0.01$, $epsilon = 0.001$, num_iterations = 10000

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
    return x**4 - 10*x + 3

# vf = np.vectorize(f)
# need to check this !!!
# https://stackoverflow.com/questions/35215161/most-efficient-way-to-map-function-over-numpy-array

x_values = list(np.arange(-5,5,0.1))
f_values = [f(x) for x in np.arange(-5,5,0.1)]

plt.figure(figsize=(6,6))

plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.plot(x_values, f_values);

```

Вычисляем производную функцию f(x):

$ f'(x) = 4x^3 - 10$
Теперь реализуем метод в виде функции:

```python
def grad_f(x):
    return 4*(x**3) - 10

def gradient_descent(x_start, learning_rate, epsilon, num_iterations):
    x_curr = x_start
    
    for _ in range(num_iterations):
        x_new = x_curr - learning_rate * grad_f(x_curr)
        
        if abs(x_new - x_curr) < epsilon:
            return x_curr

        x_curr = x_new
    
    return x_curr
        
```

Будем искать минимум с гиперпараметрами:

$x_0 = 5$, $\eta = 0.01$, $epsilon = 0.001$, num_iterations = 10000

```python
gradient_descent(5, 0.01, 0.001, 10000)
```
